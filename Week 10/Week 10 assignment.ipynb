```python
import matplotlib.pyplot as plt
```

```python
models = ['Dummy', 'KNN-1', 'KNN-2', 'KNN-3-1', 'KNN-3-2']
accuracy = [0.5075, 0.895, 0.963, 0.721, 0.965]
precision = [0.5327, 0.9045, 0.9061, 1.0, 0.8873]
recall = [0.5048, 0.8867, 0.9407, 0.4163, 0.913]
```

```python
steps = [0, 1, 2, 3, 4]
plt.figure(figsize=(12, 8))
plt.plot(steps, accuracy, 'o-', linewidth=2, label='Accuracy', color='blue')
plt.plot(steps, recall, 's-', linewidth=2, label='Recall', color='green')
plt.plot(steps, precision, '^-', linewidth=2, label='Precision', color='red')

for i in range(1, len(steps)):
    plt.axvline(x=i-0.5, color='gray', linestyle='--', alpha=0.7)
stage_positions = [-0.2, 0.8, 1.8, 2.8, 3.8]
stage_names = ['Dummy', 'KNN\nTraining', 'KNN\nPrediction', 'KNN\nRetraining', 'KNN\nRetrained\nPrediction']
for i in range(len(stage_names)):
    plt.text(stage_positions[i], 0.2, stage_names[i], ha='center', fontsize=10)
plt.title('Model Performance by Stage', fontsize=15)
plt.ylabel('Score (0-1)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.ylim(0, 1.1)
plt.xlim(-0.5, 4.5)
plt.xticks([])
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```
